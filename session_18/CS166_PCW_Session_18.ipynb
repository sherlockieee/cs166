```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Complete graph with 4 nodes
n = 4
edges = [(i, j) for i in range(n) for j in range(0, i)]
print(edges)
```

```python
g = nx.Graph(edges)
nx.draw_circular(g)
```

```python
#Complete graph with 5 nodes
n = 5
edges = [(i, j) for i in range(n) for j in range(0, i)]
g = nx.Graph(edges)
nx.draw_circular(g)
```

```python
#A bipartite graph made of a two­node group and a three­node group in which all
#possible inter­group edges are present.
n = 5
edges = [(i, j) for i in range(0, 2) for j in range(2, 5)]
g = nx.Graph(edges)
nx.draw_networkx(g, pos = nx.drawing.layout.bipartite_layout(g, [0, 1]))
```

```python
#A bipartite graph made of 2 three­node groups in which all
#possible inter­group edges are present.
n = 6
edges = [(i, j) for i in range(0, 3) for j in range(3, 6)]
g = nx.Graph(edges)
nx.draw_networkx(g, pos = nx.drawing.layout.bipartite_layout(g, [0, 1, 2]))
```

```python
#hypercube graph of 4 dimensions
g = nx.hypercube_graph(4)
nx.draw_networkx(g)
```

```python
#A “barbell graph” made of two 20­ node complete graphs 
#that are connected by a single edge.
g = nx.barbell_graph(20, 1)
nx.draw_networkx(g)
```

```python
#A “wheel graph” made of 100 nodes
g = nx.wheel_graph(100)
nx.draw_networkx(g)
```
