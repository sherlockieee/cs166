```python
import numpy as np
import matplotlib.pyplot as plt
w = o = np.linspace(0, 1, 10)

alpha = 0.03
beta = 0.3
gamma = 4

w_grid, o_grid = np.meshgrid(w, o)

vector_w = beta * w_grid * (1 - w_grid) * (1 - gamma * (o_grid))
vector_o = alpha * w_grid * o_grid

plt.figure(figsize=(6, 6))
plt.quiver(w_grid, o_grid, vector_w, vector_o)

plt.xlabel("Relationship weight")
plt.ylabel("Opinion change")
plt.show()
```
